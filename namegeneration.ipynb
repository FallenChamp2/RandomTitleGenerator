```python
import pandas as pd
import numpy as np

anime = pd.read_csv(r'../input/myanimelist/anime_cleaned.csv')
anime.head()
```

```python
anime.info()
```

```python
anime['title_english'].fillna('Ignore', inplace = True)
```

```python
from nltk import word_tokenize, pos_tag

titles = anime['title_english']

Nouns = []

for i in range (len(titles)):
    tokens = word_tokenize(titles[i])
    tagged = pos_tag(tokens)
    N = [token for token, pos in tagged if (pos == 'NN')]
    Nouns.append(N)    
```

```python
nouns = [z for z in Nouns if z]
```

```python
nouns2 = [z1 for z1 in nouns if(z1 != ['Ignore'])]
nouns2 = [z1 for z1 in nouns2 if(z1 != ['s'])]
```

```python
Adjectives = []

for i in range (len(titles)):
    tokens = word_tokenize(titles[i])
    tagged = pos_tag(tokens)
    A = [token for token, pos in tagged if (pos == 'JJ')]
    Adjectives.append(A)    
```

```python
adjectives = [z for z in Adjectives if z]
adjectives2 = [z1 for z1 in adjectives if(z1 != ['Ignore'])]
adjectives2 = [z1 for z1 in adjectives2 if(z1 != ['s'])]
```

```python
DTs = []

for i in range (len(titles)):
    tokens = word_tokenize(titles[i])
    tagged = pos_tag(tokens)
    DT = [token for token, pos in tagged if (pos == 'DT')]
    DTs.append(DT) 
```

```python
dts = [z for z in DTs if z]
dts2 = [z1 for z1 in dts if(z1 != ['Ignore'])]
```

```python
fdts = []

for s in dts2:
    for w in s:
        fdts.append(w)
```

```python
def my_function(x):
    return list(dict.fromkeys(x))

fdts2 = my_function(fdts)
#Final Determiners
```

```python
AllNouns = []

for s in nouns2:
    for w in s:
        AllNouns.append(w)
        
ANs = my_function(AllNouns)
#Final Nouns
```

```python
AllAdjectives = []

for s in adjectives2:
    for w in s:
        AllAdjectives.append(w)
        
ADJs = my_function(AllAdjectives)
#Final Adjectives
```

```python
Pronouns = []

for i in range (len(titles)):
    tokens = word_tokenize(titles[i])
    tagged = pos_tag(tokens)
    P = [token for token, pos in tagged if (pos == 'PRP')]
    Pronouns.append(P)
    
    
pronouns = [z for z in Pronouns if z]
pronouns2 = [z1 for z1 in pronouns if(z1 != ['Ignore'])]
pronouns2 = [z1 for z1 in pronouns if(z1 != ['s'])]

AllPronouns = []

for s in pronouns2:
    for w in s:
        AllPronouns.append(w)
        
Ps = my_function(AllPronouns)
```

```python
Verbs = []

for i in range (len(titles)):
    tokens = word_tokenize(titles[i])
    tagged = pos_tag(tokens)
    V = [token for token, pos in tagged if (pos == 'VB')]
    Verbs.append(V)
    
    
verbs = [z for z in Verbs if z]
verbs2 = [z1 for z1 in verbs if(z1 != ['Ignore'])]

AllVerbs = []

for s in verbs2:
    for w in s:
        AllVerbs.append(w)
        
Vs = my_function(AllVerbs)
```

```python
import itertools, random

names = []

def gen_names():
    
    for name in itertools.product(Ps[ : 3], Vs[ : 3], fdts2[ : 3], ADJs[9 : 12], ANs[9 : 12]):
        names.append(name)
        
        
full_name = []

def show_names():
    
    print('Total ' + str(len(names)) + ' Names')
    
    for n in names:
        full_name.append(' '.join(n))
        

gen_names()
show_names()

print(random.sample(full_name, 5))
```
